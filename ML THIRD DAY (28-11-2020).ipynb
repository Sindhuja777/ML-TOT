###  Merging Datasets

```python
market_df = pd.read_csv("Market_Fact.csv")
cust_df = pd.read_csv("Cust_dimen.csv")
prod_df = pd.read_csv("Prod_dim.csv")
```

```python
market_df.head(2)
```

```python
cust_df.head(2)
```

```python
prod_df.head(2)
```

```python
df1 = pd.merge(market_df,cust_df,how = 'inner',on = 'Cust_id')
df1.head(2)
```

```python
merged_df = pd.merge(df1,prod_df,how = 'inner',on = 'Prod_id')
merged_df.head()
```

## Grouping

- Grouping analysis mainly have 3 steps
    - Split the dataset into groups
    - Applying functions (like sum(columns) ..)
    - Combining result into a dataframe

```python
market_df.head()
```

* Step1: Grouping using groupby()

```python
grpby_prdid = market_df.groupby('Prod_id')
grpby_prdid
```

* Step 2: Applying Function

```python
grpby_prdid['Profit'].sum()
```

* Step 3: Converting result from series to dataframe

```python
pd.DataFrame(grpby_prdid['Profit'].sum())
```

###  Replacing the Missing Values

- Replace missing values either with '0' or mean(), id that is numerical data
- Replace missing values with 'most occured object value' or or any other object like 'No_name'
- fillna()

```python
data2 = market_df.copy()
```

```python
data2.head(2)
```

```python
data2.shape
```

```python
data2.isnull().sum()
```

```python
data2['Product_Base_Margin'].mean()
```

```python
data2['Product_Base_Margin'] = data2['Product_Base_Margin'].fillna(data2['Product_Base_Margin'].mean())
```

```python
data2.isnull().sum().sum()
```
