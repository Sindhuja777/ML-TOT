```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
duplicate = mall_data.drop('CustomerID', axis = 1)
duplicate.head()
```

```python
duplicate.isnull().sum()
```

```python
duplicate.info()
```

```python
from sklearn.preprocessing import LabelEncoder
laben = LabelEncoder()

duplicate['Genre'] = laben.fit_transform(duplicate['Genre'])
```

```python
duplicate.head()
```

```python
duplicate['Genre'].value_counts()
```

Check types of scallers and know about those.

```python
from sklearn.model_selection import train_test_split
data_train,data_test = train_test_split(duplicate, random_state = 2,test_size = 0.25)
```

```python
from sklearn.cluster import KMeans

kms = KMeans(n_clusters = 4)

kms.fit(data_train)
```
