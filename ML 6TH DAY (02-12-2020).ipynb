### Pie Chart

- plt.pie()

```python
sports = ['Cricket','Chess','football','ludo','TempleRun']
score = [80,60,75,89,98]
plt.pie(score,labels = sports,autopct = '%1.2f%%',
       colors = ['c','g','pink','r','purple'],startangle = 90,
        explode = (0,0.1,0,0,0.2),shadow = True)
plt.show()
```

```python
help(plt.pie)
```

## Area Plot

- plt.stackplot()

```python
mrng_shift = [1,2,3,4]
eng_shift = [1,4,3,1]
night_shift = [3,1,5,1]
plt.stackplot(mrng_shift,eng_shift,night_shift,colors = ['g','pink'])
plt.show()
```

### Distribution Plot

- It plots the univariable distribution of data
- sns.distplot()

```python
sns.set_style('darkgrid')
sns.distplot(iris['sepal_length'],kde = False)
```

###  Categorical Plot

- By default it returns a scatter plot
    - sns.catplot()

```python
help(sns.catplot)
```

```python
sns.catplot(x = 'species', y = 'sepal_width', data = iris)
```

```python
## It display only for numerical columns
sns.catplot(data = iris)
```

```python
sns.catplot(x = 'species', y = 'sepal_width', data = iris, palette = 'BuGn')
```

####  Boxplot

```python
sns.catplot(x = 'species', y = 'sepal_width',
            data = iris, kind = 'box')
```

```python
sns.catplot(data = iris,kind = 'box',orient = 'h',palette = 'husl')
```

### Violin Plot

- Similar to boxplot

```python
sns.catplot(x = 'species', y = 'petal_width', data = iris,kind = 'violin',palette = 'husl')
```

**Strip Plot**
- It is represented the data in sorted order

```python
sns.catplot(x = 'species', y = 'sepal_length', data = iris,kind = 'strip',palette = 'husl')
```

**Swarm Plot**
- Also known as 'Bee's Plot'

```python
sns.catplot(x = 'species', y = 'petal_length', data = iris,kind = 'swarm',palette = 'husl')
```

### Heatmaps

- Heatmap displays matrix-like graph, where all the cells are filled with the colors depeding upon the value

- sns.heatmap()

```python
t = sns.load_dataset('tips')
t.head()
```

```python
t.iloc[:,[0,1,-1]].head()
```

```python
sns.heatmap(t.iloc[:,[0,1,-1]].head(), annot = True)
```

```python
t.corr()
```

```python
sns.heatmap(t.corr(),annot = True)
```

```python
help(sns.heatmap)
```

```python
sns.heatmap(t.corr(),annot = True, cmap = 'Greens')
```
